```python
disp_alim['Disp. Alimentaire (Kcal)'] = disp_alim['Disponibilité alimentaire (Kcal/personne/jour)']*disp_alim['Population']*365*1000
disp_alim['Disp. Alimentaire (Kg)'] = disp_alim['Disponibilité alimentaire en quantité (kg/personne/an)']*disp_alim['Population']*1000
disp_alim.head()
```

```python
disp_alim['Disp. Alimentaire (Kg)'] = np.where(disp_alim['Disp. Alimentaire (Kg)']<0.01, np.nan, disp_alim['Disp. Alimentaire (Kg)'])
disp_alim['Disp. Alimentaire (Kcal)'] = np.where(disp_alim['Disp. Alimentaire (Kcal)']<0.01, np.nan, disp_alim['Disp. Alimentaire (Kcal)'])
disp_alim['Ratio énergie/poids'] = disp_alim['Disp. Alimentaire (Kcal)']/disp_alim['Disp. Alimentaire (Kg)']
disp_alim.head()
```

```python
disp_alim[disp_alim.Produit == 'Oeufs']
```

```python
disp_alim['Disp. de protéines (Kg)'] = disp_alim['Disponibilité de protéines en quantité (g/personne/jour)']*disp_alim['Population']*365
disp_alim['Disp. de protéines (Kg)'] = np.where(disp_alim['Disp. de protéines (Kg)']<0.01, np.nan, disp_alim['Disp. de protéines (Kg)'])
disp_alim['Ratio poids de protéines/poids total'] = disp_alim['Disp. de protéines (Kg)']/disp_alim['Disp. Alimentaire (Kg)']
disp_alim.head(10)
```
